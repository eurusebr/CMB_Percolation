```python
#in this piece of code we load the data and rhen we extract the exponents for the size of the gap, 
#the size of the biggest cluster befor the phase transition, and the fraction of the land where the phase transition happens
#Libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Loading Data
water1_bigcluster = np.load("/home/complex/c++/big_cluster_2.npy")
water1_totallandmass = np.load("/home/complex/c++/total_landmass_2.npy")

water2_bigcluster = np.load("/home/complex/c++/big_cluster_2.npy")
water2_totallandmass = np.load("/home/complex/c++/total_landmass_2.npy") 

water4_bigcluster = np.load("/home/complex/c++/big_cluster_4.npy")
water4_totallandmass = np.load("/home/complex/c++/total_landmass_4.npy") 

water6_bigcluster = np.load("/home/complex/c++/big_cluster_6.npy")
water6_totallandmass = np.load("/home/complex/c++/total_landmass_6.npy") 

water8_bigcluster = np.load("/home/complex/c++/big_cluster_8.npy")
water8_totallandmass = np.load("/home/complex/c++/total_landmass_8.npy") 

water11_bigcluster = np.load("/home/complex/c++/big_cluster_11.npy")
water11_totallandmass = np.load("/home/complex/c++/total_landmass_11.npy") 

water22_bigcluster = np.load("/home/complex/c++/big_cluster_22.npy")
water22_totallandmass = np.load("/home/complex/c++/total_landmass_22.npy") 
```

```python
#Calculating the exponents
critical = np.zeros((7,3))

g_subtract = []
big = water1_bigcluster
for i in range(len(big)-1):
    g_subtract.append(np.abs(big[i+1] - big[i]))

critical[0][0]= g_subtract.index(max(g_subtract)) 
critical[0][1] = max(g_subtract) 
critical[0][2] = water1_totallandmass[g_subtract.index(max(g_subtract))] 

g_subtract = []
big = water2_bigcluster
for i in range(len(big)-1):
    g_subtract.append(np.abs(big[i+1] - big[i]))

critical[1][0]= water2_bigcluster[g_subtract.index(max(g_subtract))] #The Size of the biggest cluster befor the phase transition
critical[1][1] = max(g_subtract) #Biggest Gap
critical[1][2] = water2_totallandmass[g_subtract.index(max(g_subtract))] #The Fraction of the land where the phase transition happens

g_subtract = []
big = water4_bigcluster
for i in range(len(big)-1):
    g_subtract.append(np.abs(big[i+1] - big[i]))

critical[2][0]= water4_bigcluster[g_subtract.index(max(g_subtract))]
critical[2][1] = max(g_subtract)
critical[2][2] = water4_totallandmass[g_subtract.index(max(g_subtract))]


g_subtract = []
big = water6_bigcluster
for i in range(len(big)-1):
    g_subtract.append(np.abs(big[i+1] - big[i]))

critical[3][0]= water6_bigcluster[g_subtract.index(max(g_subtract))]
critical[3][1] = max(g_subtract)
critical[3][2] = water6_totallandmass[g_subtract.index(max(g_subtract))]

g_subtract = []
big = water8_bigcluster
for i in range(len(big)-1):
    g_subtract.append(np.abs(big[i+1] - big[i]))

critical[4][0]= water8_bigcluster[g_subtract.index(max(g_subtract))]
critical[4][1] = max(g_subtract)
critical[4][2] = water8_totallandmass[g_subtract.index(max(g_subtract))]

g_subtract = []
big = water11_bigcluster
for i in range(len(big)-1):
    g_subtract.append(np.abs(big[i+1] - big[i]))
    
critical[5][0]= water11_bigcluster[g_subtract.index(max(g_subtract))]
critical[5][1] = max(g_subtract)
critical[5][2] = water11_totallandmass[g_subtract.index(max(g_subtract))]

g_subtract = []
big = water22_bigcluster
for i in range(len(big)-1):
    g_subtract.append(np.abs(big[i+1] - big[i]))
    
critical[6][0]= water22_bigcluster[g_subtract.index(max(g_subtract))]
critical[6][1] = max(g_subtract)
critical[6][2] = water22_totallandmass[g_subtract.index(max(g_subtract))]
```

```python
#Plotting gap exponent
L = np.array([21600,10800,5400,3600,2700,1964,982])
plt.style.use(['science','no-latex', 'muted'])
plt.figure(figsize=(6,4))
for i in range(7):
    plt.semilogx(L[i], critical[i][1], marker = 'D', markersize = 7, color = "blue")

m, b = np.polyfit(np.log(L), [critical[0][1], critical[1][1], critical[2][1], critical[3][1], critical[4][1], critical[5][1], critical[6][1]], 1)
plt.semilogx(L, m*np.log(L) + b, "--", linewidth=2, color = "Deepskyblue", label = "Slope = " + str(round(m, 3)))
plt.tick_params(labelsize = 14)
# plt.title("Gap Value Relative to Size", fontsize = 22)
plt.xlabel('$L$', fontsize = 22)
plt.ylabel('$g_{1}$', fontsize = 22)
plt.legend(fontsize = 14)
plt.savefig("gap_size.jpg")
```
